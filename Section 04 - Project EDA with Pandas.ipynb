```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#importing file as a dataframe
df = pd.read_csv('train.csv')
#reviewing the top 5 rows of the dataframe, use () to define number of rows you want 
df.head()
```

```python
'''
**Description of Data Frame Columns:
crim    : per capita crime rate by town.
zn      : proportion of residential land zoned for lots over 25,000 sq.ft.
indus   : proportion of non-retail business acres per town.
chas    : Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).
nox     : nitrogen oxide concentration (parts per 10 million).
rm      : average number of rooms per dwelling.
age     : proportion of owner-occupied units built prior to 1940.
dis     : weighted mean of distances to five Boston employment centers.
rad     : index of accessibility to radial highways.
tax     : full-value property-tax rate per $10,000.
ptratio : pupil-teacher ratio by town.
black   : 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.
lstat   : lower status of the population (percent).
medv    : median value of owner-occupied homes in $10000s

'''
```

```python
#reviewing the bottom 5 rows of the dataframe, use () to define number of rows you want 
df.tail()
```

```python
#pertinent information on dataset
df.info()
```

```python
# measures of centrality
print(df['crim'].mean()) #average of per capita crime rate by town
print(df['age'].median()) #avg of middle sorted value proportion of owner-occupied units built prior to 1940
print(df['ptratio'].mode()) #most common pupil-teacher ratio by town
print(df['medv'].std()) #measurement of how values are spread (closer or away) from mean
print(df['zn'].var()) #helps identify how spread the data set is from the mean
```

```python
#quick method to identify measures of central tendency for all columns of the dataset
df.describe() 
```

```python
#defining correlation, create a matrix using the corr() command
cor_matrix = df.corr()
#heatmap, (range +1 to -1) shows positive(yellow)/negative(green) correlation
plt.figure(figsize = (15,15))
sns.heatmap(cor_matrix, cmap='summer') 
```

```python
cor_matrix = df.drop(columns='ID') #since its categorical
cor_matrix = cor_matrix.corr()
```

```python
#heatmap, (range +1 to -1) shows positive(yellow)/negative(green) correlation
plt.figure(figsize = (15,15))
sns.heatmap(cor_matrix, annot=True) 
```

```python
#charles river unique values
df['chas'].unique()
```

```python
#charles river
df_nearcharles = df.loc[df['chas']==1]
df_nearcharles.sort_values('crim', ascending=False)
```

```python
df_nearcharles.describe()
```

```python
#charles river
df_awaycharles = df.loc[df['chas']==0]
df_awaycharles.sort_values('crim', ascending=False).head()
```

```python
df_awaycharles.describe()
```

# Observation
The above observations reveal thet crime is low near charles and is directly correlated with access to radial highways, inversely corelated to distance from employment centers and inversely correlated with median value of owner occupied homes.

```python
df_lowcrime = df.loc[df['crim']<=10]
df_lowcrime.sort_values('crim', ascending=False).tail()
```

```python
df_highcrime = df.loc[df['crim']>10]
df_highcrime.sort_values('crim', ascending=False).tail()
```

### Observation
The above observations reveal thet crime is corelated to distance from employment centers and inversely correlated with median value of owner occupied homes.
