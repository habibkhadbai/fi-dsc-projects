```python
#updated
### importing modules for the project
import requests
import matplotlib.pyplot as plt
import operator

### get data
macbeth = requests.get('http://www.gutenberg.org/cache/epub/2264/pg2264.txt').text

'''
If you displaying information about macbeth variable, you can take the print commands from here.
    print(type(macbeth))
    print(len(macbeth))
    print(macbeth[:500])
'''

### split the text into words & unique words
word=macbeth.split()
unique_words = list(set(word))

### creating a dictionary of unique words
uniquewordsdict = {}
for unique in unique_words:
    uniquewordsdict.update({unique : word.count(unique)})
    
sorted_uniquewordsdict = sorted(uniquewordsdict.items(), key=operator.itemgetter(1), reverse=True)
newlist=sorted_uniquewordsdict
newdict=dict(newlist[0:25])

### splitting the dictionary into unique words and counts
uniquewordslist=list(newdict.keys())
uniquewordscountlist=list(newdict.values())

### plotting graph size
plt.figure(figsize=(10,6), dpi=512, facecolor="blue")

### plotting the bar graph.
plt.bar(uniquewordslist, uniquewordscountlist, width=0.5, label='# of words', zorder=2)

### assigning labels
plt.xlabel('Unique Words',color="white")
plt.ylabel('Word Count',color="white")
plt.xticks(color="white", rotation=45)
plt.yticks(color="white", rotation="vertical")
plt.title('Macbeth Word Count',color="white")
plt.grid(which='major', axis='both', linestyle = "--", zorder=0)
plt.legend()

### displaying the graph
plt.show()
```
